## Data prepration and data cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv('Survey_results_public.csv')
df.head()
```

```python
#select only specific columns on which we want to perform analysis and form insights
df_selected=df[['Respondent' , 'Hobby' , 'Country' , 'Student' , 'Employment' , 'FormalEducation' , 'UndergradMajor' , 'CompanySize' , 'DevType' , 'YearsCoding' , 'JobSatisfaction' , 'Salary' , 'SalaryType' , 'IDE' , 'HoursComputer' , 'Gender']]
df_selected.head()
```

```python
# info about dataframe
df_selected.info()
```

```python
#shape of the data frame
df_selected.shape
# so there are 98855 rows and 16 columns
```

```python
#checking null values
df_selected.isnull().sum()
```

```python
# missing values % per column
missing_percentages = round(df_selected.isnull().sum()/len(df_selected)*100 , 2)
missing_percentages
# it seems that for salary and Salary Type have of data is missing
```

```python
#visualization of missing dadta
missing_percentages[missing_percentages !=0].plot(kind='barh')
```

```python
#so we take out the salary data for now as this will affect our data in large extend but we will inculde it later in its seprate analysis
df_selected = df_selected.drop(['Salary','SalaryType'] , axis=1)
```

```python
df_selected.head()
```

```python
#we drop missing values from the dataset
df_selected = df_selected.dropna()
```

## Exploratory data analysis

### what are the top 20 countries of respondant who filled the survey?

```python
#country data
df_selected.Country.value_counts()[:20]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plt.figure(figsize=(12,6))
#country.plot(kind='bar')
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df_selected , x='Country' , order= df_selected.Country.value_counts()[:20].index )
plt.xticks(rotation=90)
plt.show()
```

#### Most of the respondents are from united states(11393) followed by india(4597) and then united states(3088)

### what type of developer are respondents?

```python
df_selected.DevType.value_counts()
```

```python
dev = df_selected.DevType.str.split(';').apply(pd.Series , 1).stack()
percent = round(dev.value_counts()/len(dev)*100,2)
percent
```

```python
plt.figure(figsize=(12,6))
temp = dev.value_counts().index
g=sns.countplot(data=dev , x=dev.index , order = temp)
plt.xticks(rotation=90)
plt.show()
```

### are most of the respondent students

```python
students = df_selected['Student'].value_counts()
students
```

```python
labels = ['No' , 'Yes, full-time' , 'Yes,Part-time']
colors = sns.color_palette('pastel')[0:5]
plt.figure(figsize=(12,8))
plt.pie(students , labels=labels , colors=colors , autopct='%.0f%%')
plt.show()
```

### IDE used by respondents

```python
IDE=df_selected.IDE.str.split(';').apply(lambda x: pd.Series(x)).stack()
counts = IDE.value_counts()
```

### languages used by respondents

```python
lang = df['LanguageWorkedWith'].dropna()
lang=lang.str.split(';').apply(lambda x: pd.Series(x)).stack()
```

```python
lang_counts = lang.value_counts()
plt.figure(figsize=(12,12))
sns.barplot(y=lang_counts.index , x=lang_counts.values)
```
